# Preprocessing

```python
# cleanup = {"Gender" : {"M" : 0,
#                        "F":1
#                       },
#            "EDUCATION_STATUS" : {"High School" : 0,
#                                  "Graduate" : 1,
#                                  "Other" : 2
#                                 },
#            "MARITAL_STATUS" : {"Single" : 0,
#                                "Other" : 1
#                               },
#            "AGE" : {"Less than 30" : 0,
#                     "31-45" : 1,
#                     "46-65" : 2,
#                     "More than 65" : 3
#                    }
#           }
# for encoding the data given
cleanup = {"Gender" : {"M" : 0,
                       "F":1
                      },
           "EDUCATION_STATUS" : {"High School" : 0,
                                 "Graduate" : 1,
                                 "Other" : 2
                                },
           "MARITAL_STATUS" : {"Single" : 0,
                               "Other" : 1
                              },
           "AGE" : {"Less than 30" : 0,
                    "31-45" : 1,
                    "46-65" : 2,
                    "More than 65" : 3
                   },
           "PAY_JULY" : {8 : 9,
                         7 : 9,
                         6 : 9,
                         5 : 9,
                         4 : 9,
                         3 : 9,
                         2 : 9,
                         1 : 9,
                         -1 : 0,
                         -2 : 0,
                        },
           "PAY_AUG" : {8 : 9,
                         7 : 9,
                         6 : 9,
                         5 : 9,
                         4 : 9,
                         3 : 9,
                         2 : 9,
                         1 : 9,
                         -1 : 0,
                         -2 : 0,
                        },
           "PAY_SEP" : {8 : 9,
                         7 : 9,
                         6 : 9,
                         5 : 9,
                         4 : 9,
                         3 : 9,
                         2 : 9,
                         1 : 9,
                         -1 : 0,
                         -2 : 0,
                        },
           "PAY_OCT" : {8 : 9,
                         7 : 9,
                         6 : 9,
                         5 : 9,
                         4 : 9,
                         3 : 9,
                         2 : 9,
                         1 : 9,
                         -1 : 0,
                         -2 : 0,
                        },
           "PAY_NOV" : {8 : 9,
                         7 : 9,
                         6 : 9,
                         5 : 9,
                         4 : 9,
                         3 : 9,
                         2 : 9,
                         1 : 9,
                         -1 : 0,
                         -2 : 0,
                        },
           "PAY_DEC" : {8 : 9,
                         7 : 9,
                         6 : 9,
                         5 : 9,
                         4 : 9,
                         3 : 9,
                         2 : 9,
                         1 : 9,
                         -1 : 0,
                         -2 : 0,
                        }
          }
```

```python
train["Balance_Limit_V1"] = train["Balance_Limit_V1"].astype('category')
train.replace(cleanup, inplace=True)
train["Balance_Limit_V1"] = train["Balance_Limit_V1"].cat.codes
train.head(100)
```

# Model

```python
# import the libraries to train the model
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
```

```python
y=train.NEXT_MONTH_DEFAULT
```

```python
train.drop(['Client_ID'], axis=1, inplace=True)
train.drop(['NEXT_MONTH_DEFAULT'], axis=1, inplace=True)
```

```python
#split the training data into train and cross validation set
X_train, X_valid, y_train, y_valid = train_test_split(train, y,test_size=0.2,random_state=0) 
```

### Random Forest

```python
model  = RandomForestClassifier(n_estimators=200,random_state=5,verbose=1,min_samples_split=100) #set model parameters
```

```python
model.fit(X_train,y_train) #train the model
```

```python
pred = model.predict(X_valid) 
```

```python
#obtain the validation accuracy
accuracy = accuracy_score(pred,y_valid) 
accuracy
```

```python
print(confusion_matrix(pred,y_valid))
```

```python
print(classification_report(pred, y_valid)) #obtain the F1 score
```

```python
test.drop(['Client_ID'], axis=1, inplace=True)
```

```python
# cleanup the test set to make predictions
test["Balance_Limit_V1"] = test["Balance_Limit_V1"].astype('category')
test.replace(cleanup, inplace=True)
test["Balance_Limit_V1"] = test["Balance_Limit_V1"].cat.codes
test.head()      
```

```python
predictions = model.predict(test)
```

## Generate submission

```python
submission = pd.DataFrame({'Client_ID':test_['Client_ID'],'NEXT_MONTH_DEFAULT':predictions})
submission.head()
```
